# Categorical to Numerical

```python
n_valid = 12000  # same as Kaggle's test set size
n_train = len(df)-n_valid
```

```python
def split_train_val(df,n): 
    
    return df[:n].copy(), df[n:].copy()
```

```python
def preprocess_data(df, n_train):

    # Split the data
    train, val = split_train_val(df, n_train)

    # Separate features and target
    X_train = train.drop('SalePrice', axis=1)
    y_train = train['SalePrice']
    X_val = val.drop('SalePrice', axis=1)
    y_val = val['SalePrice']
    
    return X_train, X_val, y_train, y_val
```

```python
X_train, X_val, y_train, y_val = preprocess_data(df, n_train)
```

```python
# Identify categorical columns
categorical_columns = X_train.select_dtypes(include=['object']).columns
```

```python
categorical_columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Initialize a dictionary to store the LabelEncoders
label_encoders = {}
```

```python
for col in categorical_columns:
    # Initialize a LabelEncoder for each categorical column
    label_encoders[col] = LabelEncoder()
    
    # Combine unique categories from train, val, and test sets
    all_categories = pd.concat([X_train[col], X_val[col]]).unique()
    
    # Fit the encoder on all categories
    label_encoders[col].fit(all_categories)
    
    # Transform each set
    X_train[col] = label_encoders[col].transform(X_train[col])
    X_val[col] = label_encoders[col].transform(X_val[col])
    X_train[col].fillna(-1, inplace=True)
    X_val[col].fillna(-1, inplace=True)
```

```python
# Optional: Convert encoded columns to integer type
for col in categorical_columns:
    X_train[col] = X_train[col].astype(int)
    X_val[col] = X_val[col].astype(int)
```

# Datetime to Numerical Conversion

```python
def add_datepart(df, dt_name, drop=True):
    
    dt_column = df[dt_name]
    # column_dtype = dt_column.dtype
    
    df['Dateweek'] = df[dt_name].dt.isocalendar().week
    
    attr = ['year', 'month', 'day', 'dayofweek', 'dayofyear',
            'is_month_end', 'is_month_start', 'is_quarter_end', 'is_quarter_start', 'is_year_end', 'is_year_start']
    
    for a in attr: 
        df["Date" + a.capitalize()] = getattr(dt_column.dt, a) # df["saledate"].dt.year, getattr(df["saledate"].dt, "year")
        
    # df["Date" + 'Elapsed'] = dt_column.astype(np.int64) // 10 ** 9
    
    if drop: 
        df.drop(dt_name, axis=1, inplace=True)
```

# Missing Values

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
numerical_columns = X_train.select_dtypes(include=['int64', 'float64', 'int32', 'float32']).columns
```

```python
numerical_columns
```

```python
median_imputers = {}
```

```python
# Process numerical columns
for col in numerical_columns:

    if pd.isnull(X_train[col]).sum():
        # Create missing indicator for train and validation
        X_train[f'{col}_na'] = X_train[col].isna().astype(int)
        X_val[f'{col}_na'] = X_val[col].isna().astype(int)

        # Create and fit imputer on train data
        median_imputers[col] = SimpleImputer(strategy='median')
        X_train[col] = median_imputers[col].fit_transform(X_train[[col]])

        # Transform validation data
        X_val[col] = median_imputers[col].transform(X_val[[col]])

    
```

```python
# Ensure all missing indicators are present in both train and validation
all_na_columns = set(col for col in X_train.columns if col.endswith('_na'))
for col in all_na_columns:
    if col not in X_val.columns:
        X_val[col] = 0
    if col not in X_train.columns:
        X_train[col] = 0
```

```python
def print_score(m):
    
    print(f"RMSlE of train set {mean_squared_error(m.predict(X_train), y_train, squared=False)}")
    print(f"RMSlE of validation set {mean_squared_error(m.predict(X_val), y_val, squared=False)}")
    print(f"R^2 of train set {m.score(X_train, y_train)}")
    print(f"R^2 of validation set {m.score(X_val, y_val)}")
```

```python
m = RandomForestRegressor(n_estimators=70, min_samples_leaf=5, max_features=0.5, n_jobs=-1)
m.fit(X_train, y_train)
print_score(m)
```
